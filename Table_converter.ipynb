```python
import pandas as pd

def create_semester_tables(courses_dict, program):
    program_data = courses_dict.get(program, {})
    semester_tables = {}

    for semester, details in program_data.items():
        course_list = details["Courses"]
        df = pd.DataFrame(course_list)[["courseCode", "credits"]]
        semester_tables[semester] = df

    return semester_tables

semester_tables = create_semester_tables(courses, "Electrical Engineering")


for semester, table in semester_tables.items():
    print(f"\n{semester}:")
    print(table)
    print(type(table))
```
